Este notebook simplemente contiene un script que genera un archivo csv para tener en cuenta el nombre del archivo y juntarlo con su respectiva clase. 

Para eso, se importa la carpeta en donde esta el dataset completo, y se lee cada nombre de archivo, se clasifica en su respectiva clase dependiend unicamente de la primera letra (por lo que suponemos que el dataset est√° correctamente etiquetado).

```python
import os
import csv
```

```python
f = r'/Users/joaquincubelli/Desktop/Smar(t)bin/DatasetReducido/'
```

```python
with open('dataset.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    for name in os.listdir(f):
        if name[-3:]=='png':
            print(name, name[0])
            writer.writerow([name, name[0]])
    
```
